```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3, 3)
grid = plt.GridSpec(3, 3, wspace=0.4, hspace=0.3)

# 산점도
X = np.random.randn(100)
Y = np.random.randn(100)
plt.subplot(grid[0,:3]).scatter(X, Y)

# 직선 그래프
X = np.linspace(0, 4, 10)  # x축: 0부터 4까지 10개의 점 생성
Y = np.linspace(0, 4, 10)   # y축: 0부터 4까지 5개의 점 생성
plt.subplot(grid[1, :2]).plot(X, Y, color='r', linestyle='--')  # 1행 1, 2열에 그리기

# 코사인 그래프
X = np.linspace(0, 10, 100)
Y = np.cos(X)
plt.subplot(grid[2, 0]).plot(X, Y, color='g')

# 막대 그래프
X = np.arange(10)
Y = np.abs(np.random.randn(10) * 10)
plt.subplot(grid[2, 1]).bar(X, Y)

# 이미지
Z = np.random.uniform(0, 1, (10, 5))
plt.subplot(grid[1:, 2]).imshow(Z)

plt.show()
```
