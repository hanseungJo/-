```python
import numpy as np

# 3차원 배열의 크기를 지정합니다.
shape = (3, 4, 5)

# 3차원 배열을 생성합니다.
arr = np.random.randint(0, 10, shape)

# 3차원 배열을 출력합니다.
print(arr)
```

```python
import numpy as np

# 1. Shape가 (3, 3, 6)이고, 원소가 0~20 사이의 정수를 갖는 3차원 배열 생성
array_3d = np.random.randint(0, 21, (3, 3, 6))

# 2. Shape가 (5, 6)이고, 원소가 0~100 사이의 정수를 갖는 2차원 배열 생성
array_2d = np.random.randint(0, 101, (5, 6))

# 배열 출력
print("3차원 배열:")
print(array_3d)

print("\n2차원 배열:")
print(array_2d)
```

```python
import numpy as np

# 1차원 배열 생성
temp = np.arange(48)

# 1. reshape 메서드를 활용하여 (4, 3, 4) 크기의 3차원 배열 생성
array_3d = temp.reshape((4, 3, 4))

# 2. reshape 메서드를 활용하여 (6, 8) 크기의 2차원 배열 생성
array_2d = temp.reshape((6, 8))

# 배열 출력
print("3차원 배열 (4, 3, 4):")
print(array_3d)

print("\n2차원 배열 (6, 8):")
print(array_2d)

```
